```python
import yesg
import pandas as pd
dfEmpresasComparar = pd.read_excel('datasets/lista_empresas.xlsx')
dfEmpresasComparar = dfEmpresasComparar[~pd.isnull(dfEmpresasComparar.RANKING)]
```

```python
dfEmpresasComparar = pd.read_excel('datasets/lista_empresas.xlsx')
df_yahoo = None
for i, row in dfEmpresasComparar.iterrows():
    df_emp = yesg.get_historic_esg(row['Código'].replace( 'SANB4', 'SANB11') + '.SA')
    if type(df_emp) == pd.core.frame.DataFrame:
        df_emp['Nome'] = row['Nome']
        df_emp['Código'] = row['Código']
        if df_yahoo is None:
            df_yahoo = df_emp
        else:
            df_yahoo = pd.concat([df_yahoo, df_emp]).drop_duplicates()
```

```python
df_yahoo.Código.unique()
```

```python
df_yahoo['Escore_Yahoo_Norm'] = 100 - df_yahoo['Total-Score']
df_yahoo = df_yahoo.reset_index(drop=False)
df_yahoo.to_excel('datasets/escore_yahoo.xlsx', index=False)
```
