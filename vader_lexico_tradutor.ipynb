```python
import deep_translator
import string
from deep_translator import GoogleTranslator
import pandas as pd
```

```python
# separa palavras e simbolos

df = pd.read_csv('datasets/vader_lexicon.txt', sep='\t', names=['Token', 'Polarity', 'StdDev', 'Grades'])
dfWords = df[df.Token.str[0].str.isalpha()]
dfWords.to_csv('datasets/vader_lexicon_words.txt', sep='\t')

dfSymb = df[~ df.Token.str[0].str.isalpha()]
dfSymb.to_csv('datasets/vader_lexicon_symbols.txt', sep='\t')
```

```python
# traduz
gt = GoogleTranslator(source='en', target='pt')
df['translation'] = df.Token.apply(lambda x : gt.translate(x) if x[0].isalpha() else x)
df
```
